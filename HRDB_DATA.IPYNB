4.3 Work-Life Balance and Satisfaction
- Evaluate job satisfaction and work-life balance ratings for employees who
left vs. those who stayed.
- Determine if employees with lower work-life balance scores are more likely
to leave.
- Assess environmental and relationship satisfaction to understand their impact

Determine is employees with lower work life balance scores are more likely to leave . A measure of work-life balance satisfaction is: 1 =
Bad, 4 = Best..   Attrition: Indicates whether the employee left the company (Yes/No). Show me a rellevant explication in words and a rellevant professional graph 

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Bar chart: Number of Employees in each Work-Life Balance category by Attrition status
plt.figure(figsize=(8,5))
ax = sns.countplot(data=df, x='WorkLifeBalance', hue='Attrition', palette='mako')

# Add value labels for each bar
for c in ax.containers:
    labels = [f'{v.get_height()}' if v.get_height() > 0 else '' for v in c]
    ax.bar_label(c, labels=labels, label_type='edge', fontsize=10)

plt.title('Work-Life Balance Distribution by Attrition Status')
plt.xlabel('Work-Life Balance (1=Bad, 4=Best)')
plt.ylabel('Number of Employees')
plt.legend(title='Attrition')
plt.tight_layout()
plt.show()

# Calculate distribution: % of 'Yes' and % of 'No' per WorkLifeBalance, sum to 100% for each score
wlb_counts = df.groupby(['WorkLifeBalance', 'Attrition']).size().unstack(fill_value=0)
wlb_percent_each = wlb_counts.div(wlb_counts.sum(axis=1), axis=0) * 100

# Stacked bar chart (percent stacked) - each WLB score sums to 100%
plt.figure(figsize=(7,4))
attrition_colors = {'No': '#77b7c5', 'Yes': '#e78a8a'}
bottom = None
for attr in ['No', 'Yes']:
    val = wlb_percent_each[attr]
    plt.bar(wlb_percent_each.index, val, bottom=bottom, label=f'Attrition: {attr}', color=attrition_colors[attr])
    if bottom is None:
        bottom = val
    else:
        bottom = bottom + val

# Add percentage value labels to each bar section (percent in segment, not cumulative)
for i, wlb in enumerate(wlb_percent_each.index):
    y0 = wlb_percent_each.loc[wlb, 'No']
    y1 = wlb_percent_each.loc[wlb, 'Yes']
    plt.text(wlb, y0/2, f'{y0:.1f}%', ha='center', va='center', color='white', fontsize=10)
    plt.text(wlb, y0 + y1/2, f'{y1:.1f}%', ha='center', va='center', color='black', fontsize=10)

plt.xlabel('Work-Life Balance Score (1=Bad, 4=Best)')
plt.ylabel('% of Employees')
plt.title('Employee Distribution by Work-Life Balance and Attrition (%)')
plt.ylim(0, 100)
plt.legend(title="Attrition", bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
plt.tight_layout()
plt.show()

# In words:
print(
    "Explanation:\n"
    "The first chart displays the number of employees for each Work-Life Balance score, separated by those who left (Attrition=Yes) and those who stayed (Attrition=No). "
    "Each bar is labeled with the raw count.\n\n"
    "The second chart visualizes the percentage of employees that stayed or left for each Work-Life Balance score, "
    "with each score's 'Yes' + 'No' adding up to 100%. The chart clearly shows the proportion of leavers in each group. "
    "Darker bars indicate the percentage who left. As shown, employees with low Work-Life Balance (WLB scores 1 or 2) "
    "have a noticeably higher percentage of attrition compared to those with the best scores (3 or 4).\n\n"
    "This emphasizes: among employees within each Work-Life Balance score, those with lower scores have a much higher risk of leaving, "
    "suggesting that Work-Life Balance is a key predictor for attrition."
)
```

Explanation:
The first chart shows the distribution of employees with each Work-Life Balance score, split by whether they left the company (Attrition=Yes) or stayed (Attrition=No).

The second chart presents the attrition rate (percentage of employees who left) for each Work-Life Balance score.

From these charts, we can observe that employees with lower Work-Life Balance scores (1=Bad, 2) generally have higher attrition rates, whereas those with better Work-Life Balance (scores 3 and 4) are less likely to leave the company. This suggests a clear correlation: employees who rate their work-life balance lower are indeed more likely to leave.

```python
# Calculate attrition counts and rates
attrition_counts = df['Attrition'].value_counts()
attrition_rate = attrition_counts['Yes'] / attrition_counts.sum() * 100

# Bar chart: Number of employees by Attrition
plt.figure(figsize=(5,4))
bars = plt.bar(attrition_counts.index, attrition_counts.values, color=['#8fd9b6','#e78a8a'])

for bar, val in zip(bars, attrition_counts.values):
    plt.text(bar.get_x() + bar.get_width()/2., bar.get_height() + 5, f'{val}', ha='center', fontsize=12)

plt.xlabel('Attrition')
plt.ylabel('Number of Employees')
plt.title('Employee Attrition Count')
plt.tight_layout()
plt.show()

# Pie chart: Attrition rate in percentage
plt.figure(figsize=(5,4))
labels = ['Stayed', 'Left']
sizes = [100-attrition_rate, attrition_rate]
colors = ['#8fd9b6','#e78a8a']
explode = (0, 0.08)
plt.pie(
    sizes, labels=labels, autopct='%1.1f%%', colors=colors,
    startangle=90, explode=explode, shadow=True, textprops={'fontsize': 12}
)
plt.title('Overall Attrition Rate (%)')
plt.tight_layout()
plt.show()
```

```python
# Calculate attrition rate for each Job Satisfaction score
job_satisfaction_attrition = (
    df[df["Attrition"] == "Yes"]
    .groupby("JobSatisfaction")
    .size() / df.groupby("JobSatisfaction").size()
) * 100

# Create the barplot with uniform color
plt.figure(figsize=(7,5))
bars = plt.bar(
    job_satisfaction_attrition.index.astype(str),
    job_satisfaction_attrition.values,
    color="#8fd9b6"  # uniform color for all columns
)
for bar, val in zip(bars, job_satisfaction_attrition.values):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.7, f"{val:.1f}%", ha="center", fontsize=12)

plt.xlabel("Job Satisfaction (1=Low, 4=High)")
plt.ylabel("Attrition Rate (%)")
plt.title("Attrition Rate by Job Satisfaction")
plt.ylim(0, job_satisfaction_attrition.max() + 5)
plt.grid(axis="y", linestyle="--", alpha=0.3)
plt.tight_layout()
plt.show()
```
